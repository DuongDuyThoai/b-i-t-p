```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
food=ctrl.Antecedent(np.arange(0,11,1),'food')
service=ctrl.Antecedent(np.arange(0,11,1),'service')
tip=ctrl.Consequent(np.arange(10,26,1),'tip')
food['poor']=fuzz.trimf(food.universe,[0,0,5])
food['average']=fuzz.trimf(food.universe,[0,5,10])
food['good']=fuzz.trimf(food.universe,[5,10,10])
service['poor']=fuzz.trimf(service.universe,[0,0,5])
service['average']=fuzz.trimf(service.universe,[0,5,10])
service['good']=fuzz.trimf(service.universe,[5,10,10])
tip['less']=fuzz.trimf(tip.universe,[10,14,18])
tip['normal']=fuzz.trimf(tip.universe,[14,18,22])
tip['much']=fuzz.trimf(tip.universe,[18,22,25])
food.view()
service.view()
tip.view()
```

```python
rule1=ctrl.Rule(food['poor']&service['poor'],tip['less'])
rule2=ctrl.Rule(food['average']&service['average'],tip['normal'])
rule3=ctrl.Rule(food['good']&service['good'],tip['much'])
rule4=ctrl.Rule(food['poor']&service['average'],tip['normal'])
rule5=ctrl.Rule(food['poor']&service['good'],tip['normal'])
rule6=ctrl.Rule(food['average']&service['poor'],tip['normal'])
rule7=ctrl.Rule(food['good']&service['poor'],tip['normal'])
rule8=ctrl.Rule(food['average']&service['good'],tip['normal'])
rule9=ctrl.Rule(food['good']&service['average'],tip['normal'])
tipping_ctrl=ctrl.ControlSystem([rule1,rule2,rule3,rule4,rule5,rule6,rule7,rule8,rule9])
tipping=ctrl.ControlSystemSimulation(tipping_ctrl)
tipping.input['food']=6.2
tipping.input['service']=7.5
tipping.compute()
print(tipping.output['tip'])
tip.view(sim=tipping)
                
```

```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
time=ctrl.Antecedent(np.arange(10,120,1),'time')
rice=ctrl.Antecedent(np.arange(100,1000,1),'rice')
power=ctrl.Consequent(np.arange(0,100,1),'power')

time['rat nhanh']=fuzz.trimf(time.universe,[10,10,30])
time['nhanh']=fuzz.trimf(time.universe,[10,30,60])
time['trung binh']=fuzz.trimf(time.universe,[30,60,90])
time['cham']=fuzz.trimf(time.universe,[60,90,120])
time['rat cham']=fuzz.trimf(time.universe,[90,120,120])

rice['it']=fuzz.trimf(rice.universe,[100,100,400])
rice['trung binh']=fuzz.trimf(rice.universe,[400,550,700])
rice['nhieu']=fuzz.trimf(rice.universe,[700,1000,1000])

power['sieu tiet kiem']=fuzz.trimf(power.universe,[0,0,25])
power['tiet kiem']=fuzz.trimf(power.universe,[0,25,50])
power['binh thuong']=fuzz.trimf(power.universe,[25,50,75])
power['cao']=fuzz.trimf(power.universe,[50,75,100])
power['sieu cao']=fuzz.trimf(power.universe,[75,100,100])

time.view()
rice.view()
power.view()

rule1=ctrl.Rule(time['rat nhanh']&rice['it'],power['binh thuong'])
rule2=ctrl.Rule(time['nhanh']&rice['it'],power['binh thuong'])
rule3=ctrl.Rule(time['trung binh']&rice['it'],power['tiet kiem'])
rule4=ctrl.Rule(time['cham']&rice['it'],power['tiet kiem'])
rule5=ctrl.Rule(time['rat cham']&rice['it'],power['sieu tiet kiem'])
rule6=ctrl.Rule(time['rat nhanh']&rice['trung binh'],power['sieu cao'])
rule7=ctrl.Rule(time['nhanh']&rice['trung binh'],power['cao'])
rule8=ctrl.Rule(time['trung binh']&rice['trung binh'],power['binh thuong'])
rule9=ctrl.Rule(time['cham']&rice['trung binh'],power['binh thuong'])
rule10=ctrl.Rule(time['rat cham']&rice['trung binh'],power['tiet kiem'])
rule11=ctrl.Rule(time['rat nhanh']&rice['nhieu'],power['sieu cao'])
rule12=ctrl.Rule(time['nhanh']&rice['nhieu'],power['sieu cao'])
rule13=ctrl.Rule(time['trung binh']&rice['nhieu'],power['cao'])
rule14=ctrl.Rule(time['cham']&rice['nhieu'],power['binh thuong'])
rule15=ctrl.Rule(time['rat cham']&rice['nhieu'],power['tiet kiem'])

cooker_ctrl=ctrl.ControlSystem([rule1,rule2,rule3,rule4,rule5,rule6,rule7,rule8,rule9,rule10,rule11,rule12,rule13,rule14,rule15])
cooker=ctrl.ControlSystemSimulation(cooker_ctrl)
cooker.input['time']=30
cooker.input['rice']=500
cooker.compute()
print(cooker.output['power'])
power.view(sim=cooker)
```
